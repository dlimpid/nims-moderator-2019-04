```python
import importlib
from pkg_resources import parse_version

requirements = {
    'scipy': '1.2',
    'numpy': '1.16',
    'pandas': '0.24',
    'sklearn': '0.20',
    'matplotlib': '3',
    'seaborn': '0.9',
    'PIL': '5',
    'IPython': '6.5',
}

for package, required_version in requirements.items():
    try:
        i = importlib.import_module(package)
        if parse_version(i.__version__) < parse_version(required_version):
            raise ImportError
        print(f'{package}: {i.__version__}')
    except ImportError:
        print(f'ERROR: Cannot import {package} >= {required_version}')
```
